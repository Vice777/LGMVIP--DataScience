```python
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras import Sequential , callbacks ,Model
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, BatchNormalization , LSTM 
from tensorflow.keras.layers import Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam

from tensorflow.keras.datasets import mnist
```

```python
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from mlxtend.evaluate import scoring
```

```python
(X_train , y_train),(X_test, y_test) = mnist.load_data()
```

```python
n_images = 9
fig, axes = plt.subplots(3, 3 , figsize=(20,20))
indexes = np.random.randint(0,X_train.shape[0] , n_images)
count=0
for i in range(3):
    for j in range(3):
        axes[i][j].imshow(X_train[indexes[count]] , cmap='gray')
        axes[i][j].set_title(y_train[indexes[count]])
        count += 1
```

```python
def build_model():
    inputs = Input(shape=(28,28,1))
    x = Flatten()(inputs)
    x = Dense(128 , activation='relu')(x)
    x = Dense(64 , activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(32 , activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(10, activation='softmax')(x)
    
    model = Model(inputs=[inputs] , outputs =[outputs], name='my_model')
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
X_test = tf.expand_dims(X_test / 255.0 , -1)
X_train = tf.expand_dims(X_train / 255.0 , -1)
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.0025) ,  metrics = ['accuracy'])
history = model.fit(X_train,y_train,epochs=10 , validation_data=(X_test,y_test) , batch_size=64)
```

```python
y_test.shape
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
Y_pred = []

for i in range(10000):
    Y_pred.append(y_pred[i].argmax())
```

```python
Y_Pred_model = np.array(Y_pred)
Y_Pred_model
```

```python
Y_Pred_model
```

```python
mat = confusion_matrix(y_test,Y_Pred_model)
plot_confusion_matrix(conf_mat=mat  , show_normed=False,figsize=(6,6))
```

```python
res = scoring(y_target=y_test, y_predicted=Y_Pred_model, metric='accuracy')
print('Accuracy(%) : ' , (res * 100))
```
